```python
import sympy as sp
import numpy as np
from matplotlib import pyplot as plt
```

```python
def demand_vinter(Qv):
    return 7500-12.5*Qv
```

```python
def marginalrevenueV(Qv):
    return 7500 - 25*Qv
```

```python
Qv=sp.symbols('Qv', real=True, positive=True)
equ=sp.Eq(marginalrevenueV(Qv),800)
equ
```

```python
# Optimalt antall hotellrom solgt på vinteren
Qv_equ=sp.solve(equ,Qv)[0]
Qv_equ
```

```python
# optimal pris for rom på vinteren:
pv_eq=demand_vinter(Qv_equ)
pv_eq
```

```python
# Samfunnsøkonomisk overkudd
Pv_max = 7500
Pv_optimal = 4150
Qv_optimal = 268
MC = 800
Ps_max = 2000
Ps_optimal = 1400
Qs_optimal =240

# Funksjoner for å beregne KO, PO, og SO
def konsumentoverskudd(P_max, P_optimal, Q_optimal):
    return 0.5 * (P_max - P_optimal) * Q_optimal

def produsentoverskudd(P_optimal, MC, Q_optimal):
    return (P_optimal - MC) * Q_optimal

def samfunnsoekonomisk_overskudd(KO, PO):
    return KO + PO

# Beregner KO, PO, og SO for vinter- og sommersesongen
KO_vinter = konsumentoverskudd(Pv_max, Pv_optimal, Qv_optimal)
PO_vinter = produsentoverskudd(Pv_optimal, MC, Qv_optimal)
SO_vinter = samfunnsoekonomisk_overskudd(KO_vinter, PO_vinter)

KO_sommer = konsumentoverskudd(Ps_max, Ps_optimal, Qs_optimal)
PO_sommer = produsentoverskudd(Ps_optimal, MC, Qs_optimal)
SO_sommer = samfunnsoekonomisk_overskudd(KO_sommer, PO_sommer)

SO_total = SO_vinter + SO_sommer

# Utskrift av resultatene
print(f"Konsumentoverskudd Vinter: {KO_vinter:.2f} kr, Sommer: {KO_sommer:.2f} kr")
print(f"Produsentoverskudd Vinter: {PO_vinter:.2f} kr, Sommer: {PO_sommer:.2f} kr")
print(f"Samfunnsøkonomisk Overskudd Vinter: {SO_vinter:.2f} kr, Sommer: {SO_sommer:.2f} kr, Totalt: {SO_total:.2f} kr")
```
